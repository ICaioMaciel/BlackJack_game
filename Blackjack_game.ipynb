```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')

ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')

values = {
        'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6,
        'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
        'Queen':10, 'King':10, 'Ace':11
     }

playing = True
```

```python
class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit, rank))
    
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += '\n' + card.__str__()
        return "The deck has: " + deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self, card):
        self.cards.append(card)
        self.value += values[card.rank]
        
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips:
    def __init__(self, total = 100):
        self.total = 100
        self.bet = 0
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet? '))
        except ValueError:
            print('Sorry, a bet must be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)
            else:
                break
```

```python
def hit(deck, hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck, hand):
    global playing
    
    while True:
        x = input("Would you like to Hit or Stand? enter H or S").lower()
        
        if x[0] == 'h':
            hit(deck, hand)
        elif x == 's':
            print('Player Stands. Dealer is Playing')
            playing = False
        else:
            print("Sorry try again")
            continue
        break
```

```python
def show_some(player, dealer):
    
    print("\nDealers's hand: ")
    print(f"First card: {dealer.cards[0]}")
    print("Second card is hidden!")
    
    
    print("\nPlayer`s hand:")
    for card in player.cards:
        print(card)
    
def show_all(player, dealer):

    print(f"Value of dealer's cards: {dealer.value}")
    print("\n Dealer`s hand:")
    for card in dealer.cards:
        print(card)
    
    print(f"Value of player's cards: {player.value}")
    print("\nPlayer`s hand:")
    for card in player.cards:
        print(card)
```

```python
def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    chips.lose_bet()
    
def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
```
